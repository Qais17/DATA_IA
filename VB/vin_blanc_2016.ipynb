```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
vin = pd.read_excel ("2016-stats-recolte.xls")
vini = vin.dropna()

vini.columns = ["description", "nombre de declarations de recolte et de production souscrites (ha)", "superficie totale des vignes en production (ha)", "A.O.P (ha)", "aptes au Cognac ou Armagnac (ha)", "IGP (ha)", "VSIG (ha)", "A.O.P blanc (hl)", "rouge (hl)", "rosé (hl)", "VSI/VCI (hl)", "I.G.P blanc (hl)", "rouge (hl)", "rosé (hl)", "VSIG blanc (hl)", "rouge (hl)", "rosé (hl)", "quantites de vins aptes a la production de Cognac ou d'Armagnac (hl)", "production commercialisable : Volumes AOP, IGP, VSIG et vins aptes (hl)", "production non commercialisable (hl)", "total (hl)"]
vini1 = vini.set_index("description")

vini1
```

```python
WT = vini1.drop(index = "TOTAUX", columns='total (hl)')

DF_blanc = WT['A.O.P blanc (hl)']

DF_blanc.sort_values(ascending = False)

blanc_max = DF_blanc[DF_blanc > 100000]

blanc_max.shape

blanc_min = DF_blanc.sort_values(ascending = False).iloc[19:79]

blanc_min = blanc_min.sum()

blanc_max = blanc_max.reset_index()

blanc_max.loc[len(blanc_max)]=['AUTRES', blanc_min]

blanc_max.set_index("description")
```

```python
fig, ax = plt.subplots(figsize = (15,5))

blanc_max.plot.bar(x = "description", y='A.O.P blanc (hl)', title = "Production de vin blanc A.O.P par région en 2016", ax=ax)

```
