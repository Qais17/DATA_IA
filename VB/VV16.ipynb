```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas import DataFrame
from pandas.tools.plotting import table
import matplotlib.cbook as cbook
vv16 = pd.read_excel('data/re16pro.xls')
vv16
```

```python
plt.pie(
    # using data total)arrests
    vv16['Sup vignes cognac - armagniac (ha)'],
    # with the labels being officer names
    labels=vv16['DEPARTEMENTS'],
    startangle=100,
    # with the percent listed as a fraction
    autopct='%1.1f%%', 
    )

# View the plot drop above
plt.axis('equal')
# View the plot
plt.tight_layout()
plt.show()
# en 2016 aussi, ont remarque que pour Sup vignes cognac - armagniac (ha) c'est les regions de CHARENTE 
# ET CHARENTE-MAR qui ont le plus d'hectar 
```

```python
da = vv16.set_index('DEPARTEMENTS').loc[:,['Sup vignes cognac - armagniac (ha)', 'Qt Cognac ou Armagnec (hl)']]
da
```

```python
classement = da.sort_values(['Sup vignes cognac - armagniac (ha)', 'Qt Cognac ou Armagnec (hl)'], ascending=False)
classement
```

```python
top = classement.iloc[:2]
top
```

```python
autre = classement.iloc[3:90]
autre
```

```python
autres = autre.sum()
autres
```

```python
top = classement.iloc[:2]
top = top.reset_index()
perfect = top.append(autres.T, ignore_index=True).fillna("AUTRE").set_index("DEPARTEMENTS")
perfect
```

```python
plt.pie(
    # using data total)arrests
    perfect['Sup vignes cognac - armagniac (ha)'],
    # with the labels being officer names
    labels=perfect.index,
    startangle=100,
    # with the percent listed as a fraction
    autopct='%1.1f%%', 
    )

# View the plot drop above
plt.axis('equal')
# View the plot
plt.tight_layout()
plt.show()
```
